```python
import random

def password(length):
    pwd = str()
    characters = "abczxyopqdethsu" + "0123456789"
    for i in range(length):
        pwd = pwd + random.choice(characters)
    return pwd
```

```python
class NewList(list):
    def remove_max(self):
        self.remove(max(self))
    def append_sum(self):
        self.append(sum(self))

x = NewList([1,2,3])
while max(x) < 10:
    x.remove_max()
    x.append_sum()

print(x)

'''
Upon first removing the maximal element (3 ) and adding then sum of the MyList object (3 ), the object has the same elements in the same locations as before ([1, 2, 3] ). 
Therefore, the while condition will always be met, and the program will run as long as it can.
'''
```

# Customizing Your Plots

```python
x = np.linspace(0,10,20)
y1 = x ** 2.0
y2 = x ** 1.5
plt.plot(x, y1, "bo-", linewidth = 2, markersize=12, label="First")
plt.plot(x, y2, "gs-", linewidth = 2, markersize=12, label="Second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("myplot.pdf")
```

## tic-tac-toe board, we will use a numpy array with dimension 3 by 3.

```python
import numpy as np
import random

def create_board():
    board = np.zeros((3,3), dtype=int)
    return board
```

```python
def place(board, player, position):
    if board[position] == 0:
        board[position] = player
        return board

board = create_board()
place(board, 1, (0, 0))
```

```python
def possibilities(board):
    return list(zip(*np.where(board == 0)))

possibilities(board)
#[(0, 1), (0, 2), (1, 0), (1, 1), (1, 2), (2, 0), (2, 1), (2, 2)]
```

```python
random.seed(1)

def random_place(board, player):
    selections = possibilities(board)
    #print(f'Selection : {selections}')
    if len(selections) > 0:
        selection = random.choice(selections)
        place(board, player, selection)
    return board

random_place(board, 2)
# array([[1, 0, 0],
#        [2, 0, 0],
#        [0, 0, 0]])
```

```python
random.seed(1)
board = create_board()

for i in range(3):
    for player in [1, 2]:
        random_place(board, player)
```

```python
def row_win(board, player):
    if np.any(np.all(board==player, axis=1)): # this checks if any row contains all positions equal to player.
        return True
    else:
        return False

row_win(board, 1)
```

```python
def col_win(board, player):
    if np.any(np.all(board==player, axis=0)): # this checks if any col contains all positions equal to player.
        return True
    else:
        return False

row_win(board, 1)
```

```python
def diag_win(board, player):
    if np.all(np.diag(board)==player) or np.all(np.diag(np.fliplr(board))==player):
        # np.diag returns the diagonal of the array
        # np.fliplr rearranges columns in reverse order
        return True
    else:
        return False
    
board[1,1] = 2
diag_win(board, 2)
```
